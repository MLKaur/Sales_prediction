**LOAD DATASET**

```python
dataset = pd.read_csv('DigitalAd_dataset.csv')
```

**SUMMARISE DATASET**

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
print("\n🔹 Dataset Info:")
dataset.info()

print("\n🔹 Statistical Summary:")
display(dataset.describe())

print("\n🔹 Missing Values in Each Column:")
print(dataset.isnull().sum())

print("\n🔹 Number of duplicate rows:", dataset.duplicated().sum())

print("\n🔹 Unique Values in Each Column:")
for col in dataset.columns:
    print(f"{col}: {dataset[col].nunique()}")

print("\n🔹 Value Counts:")
for col in dataset.select_dtypes(include='object').columns:
    print(f"\n{col}:\n{dataset[col].value_counts()}")

if 'Purchased' in dataset.columns:
    print("\n🔹 Target Variable Distribution:")
    sns.countplot(data=dataset, x='Purchased')
    plt.title("Target Variable: Purchased")
    plt.show()

dataset.hist(bins=30, figsize=(12, 8), color='skyblue')
plt.suptitle("Histogram of Numerical Features")
plt.show()

if dataset.select_dtypes(include=np.number).shape[1] > 1:
    plt.figure(figsize=(8, 6))
    sns.heatmap(dataset.corr(), annot=True, cmap='viridis')
    plt.title("Correlation Heatmap")
    plt.show()

numerical_features = dataset.select_dtypes(include=np.number).columns.tolist()
for feature in numerical_features:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=dataset[feature])
    plt.title(f"Boxplot of {feature}")
    plt.show()
```

**SEGREGATE DATASET INTO INPUT AND OUTPUT VARIABLE**

```python
X = dataset.iloc[:, :-1].values
X
```

```python
Y = dataset.iloc[:, -1].values
Y
```

**SPLITTING INTO TRAIN AND TEST**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

**FEATURE SCALING**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**TRAINING**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train, y_train)
```

**Predicting, wheather new customer with Age & Salary will Buy or Not**

```python
age = int(input("Enter New Customer Age: "))
sal = int(input("Enter New Customer Salary: "))
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print("Customer will Buy")
else:
  print("Customer won't Buy")
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

**EVALUATION USING CONFUSION MATRIX AND ACCURACY**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
